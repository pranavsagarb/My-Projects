```python
# Boto3 is the Amazon Web Services (AWS) Software Development Kit (SDK) for Python

import sagemaker
import boto3

# creating a Sagemaker session
sagemaker_session = sagemaker.Session()

bucket = 'elitecloud-project'  # bucket name
prefix = 'pca'  # prefix is the subfolder within the bucket

role = sagemaker.get_execution_role()
```

```python
import io
import numpy as np
import sagemaker.amazon.common as smac

buf = io.BytesIO() # create an in-memory byte array
df_matrix = dataframe_final.to_numpy() # convert the dataframe into 2-dimensional array
smac.write_numpy_to_dense_tensor(buf, df_matrix)
buf.seek(0)
```

```python
import os
key = 'pca'

#following code uploads the data in record-io format to S3 bucket
boto3.resource('s3').Bucket(bucket).Object(os.path.join(prefix, 'train', key)).upload_fileobj(buf)

# It prints out the training data location in s3
s3_train_data = 's3://{}/{}/train/{}'.format(bucket, prefix, key)

print('uploaded training data location: {}'.format(s3_train_data))
```

```python
output_location = 's3://{}/{}/output'.format(bucket, prefix)
print('training artifacts will be uploaded to: {}'.format(output_location))
```

```python
# The below code obtains a reference to the pca container image

from sagemaker.amazon.amazon_estimator import get_image_uri

container = get_image_uri(boto3.Session().region_name, 'pca')
```

```python
# We have pass in the container, the type of instance that we would like to use for training 
# output path and sagemaker session into the Estimator. 
# Specified how many instances we would like to use for training here


pca = sagemaker.estimator.Estimator(container,
                                       role, 
                                       train_instance_count=1, 
                                       train_instance_type='ml.c4.xlarge',
                                       output_path=output_location,
                                       sagemaker_session=sagemaker_session)


pca.set_hyperparameters(feature_dim=11,
                        num_components=6,
                        subtract_mean=False,
                        algorithm_mode='regular',
                        mini_batch_size=100)


# Pass in the training data from S3 to train the pca model


pca.fit({'train': s3_train_data})

# Let's see the progress using cloudwatch logs
```

```python
# Deploying the trained PCA model
```

```python
# Deploy the model to perform inference 

pca_reduction = pca.deploy(initial_instance_count = 1,
                                          instance_type = 'ml.m4.xlarge')
```

```python
from sagemaker.predictor import csv_serializer, json_deserializer

# Here the Deserializer accepts two arguments, the result data and the response content type, and return a sequence of bytes in the specified content type.

#pca_reduction.content_type = 'text/csv'
pca_reduction.serializer = csv_serializer
pca_reduction.deserializer = json_deserializer
```

```python
# prediction on the test data

result = pca_reduction.predict(np.array(dataframe_final))
```

```python
container = get_image_uri(boto3.Session().region_name, 'xgboost','0.90-2')
```

```python
Xgboost_classifier = sagemaker.estimator.Estimator(container,
                                       role, 
                                       train_instance_count=1, 
                                       train_instance_type='ml.m4.xlarge',
                                       output_path=output_location,
                                       sagemaker_session=sagemaker_session)

# To improve the performance of the model hyperparameters tuning is done 

Xgboost_classifier.set_hyperparameters(max_depth=8,
                           objective='multi:softmax',
                           num_class= 2,
                           eta = 0.5,
                           num_round = 150
                           )
```

```python
train_input = sagemaker.session.s3_input(s3_data = s3_train_data, content_type='csv',s3_data_type = 'S3Prefix')
valid_input = sagemaker.session.s3_input(s3_data = s3_valid_data, content_type='csv',s3_data_type = 'S3Prefix')

Xgboost_classifier.fit({'train': train_input, 'validation': valid_input})
```

```python
# convert the values in bytes format to array

def bytes_2_array(x):
    
    l = str(x).split(',')
    
    l[0] = l[0][2:]
    
    l[-1] = l[-1][:-1]
    
    #iterating through the list of strings and converting them into float type
    for i in range(len(l)):
        l[i] = float(l[i])
        
    #converting the list to into array
    l = np.array(l).astype('float32')
    
    #reshape 1-d array to 2-d array
    return l.reshape(-1,1)
```

```python
predicted_values = bytes_2_array(XGB_prediction)
```
